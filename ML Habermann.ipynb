```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
haberman=pd.read_csv('Downloads/haberman.csv',names=['Age','year','nodes','status'])
```

```python
haberman1['status']=haberman1['status'].map({1:True,2:False})
```

```python
haberman1.plot(kind="scatter",x="Age",y="nodes")
plt.show()
```

```python
haberman1=haberman1.astype(int)
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(haberman1,hue="status",size=5).map(plt.scatter,"Age","nodes").add_legend();
plt.show();
```

```python
plt.close();
sns.set_style("whitegrid");
sns.pairplot(haberman1,hue="status",size=3);
plt.show();
```

```python
haberman_1=haberman1.loc[haberman1["status"]==1]
haberman_2=haberman1.loc[haberman1["status"]==2]
```

```python
plt.plot(haberman_1["Age"],np.zeros_like(haberman_1["Age"]),"o")
plt.plot(haberman_2["Age"],np.zeros_like(haberman_2["Age"]),"o")
```

```python
sns.FacetGrid(haberman1,hue="status",size=3).map(sns.distplot,"Age").add_legend();
plt.show();
```

```python
sns.FacetGrid(haberman1, hue="status", size=5) \
   .map(sns.distplot, "year") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(haberman1, hue="status", size=5) \
   .map(sns.distplot, "nodes") \
   .add_legend();
plt.show();
```
